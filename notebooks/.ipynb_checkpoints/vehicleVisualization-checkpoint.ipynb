```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

vehicles = pd.read_csv("http://knuth.luther.edu/~osmaah02/cars.csv")
```

```python
vehicles.price = vehicles.price[~((vehicles.price-vehicles.price.mean()).abs() > .05*vehicles.price.std())]
vehicles.odometer = vehicles.odometer[~((vehicles.odometer-vehicles.odometer.mean()).abs() > 5*vehicles.odometer.std())]
vehicles.year = vehicles.year[~((vehicles.year-vehicles.year.mean()).abs() > 10*vehicles.year.std())]
vehicles.state_name = vehicles.state_name.replace({"FAILED": ""})

vehicles.dtypes
```

```python
vehicles.plot.scatter(x="odometer", y="price", c="year", colormap="coolwarm", figsize=(16,10))
plt.show()
vehicles.price.dropna().corr(vehicles.weather.dropna())
print("This looks stupid")
```

```python
fig, axes = plt.subplots(nrows = 4, figsize=(20, 60))
nicePrices = vehicles['price'].mask(vehicles['price'] > 100000, np.nan)
axes[0].violinplot(nicePrices.dropna(), showmeans = True)

nicePrices = vehicles["price"].mask(vehicles["price"] > 50000, np.nan)
axes[1].violinplot(nicePrices.dropna(), showmeans = True)

nicePrices = vehicles['price'].mask(vehicles['price'] > 25000, np.nan)
axes[2].violinplot(nicePrices.dropna(), showmeans = True)

nicePrices = vehicles['price'].mask(vehicles['price'] > 10000, np.nan)
axes[3].violinplot(nicePrices.dropna(), showmeans = True)

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 4, figsize=(20, 60))
axes[0].violinplot(vehicles.odometer.dropna(), showmeans = True)

niceOdom = vehicles['odometer'].mask(vehicles['odometer'] > 500000, np.nan)
axes[1].violinplot(niceOdom.dropna(), showmeans = True)

niceOdom = vehicles['odometer'].mask(vehicles['odometer'] > 300000, np.nan)
axes[2].violinplot(niceOdom.dropna(), showmeans = True)

niceOdom = vehicles['odometer'].mask(vehicles['odometer'] > 200000, np.nan)
axes[3].violinplot(niceOdom.dropna(), showmeans = True)

plt.show()
```

```python
rawConditions = vehicles.condition.value_counts()
# quantiles = [0]
# for i in range(10):
#     quantiles.append(vehicles.odometer.quantile((i+1)/10))
conditions = []
for i in rawConditions.iteritems():
    conditions.append(i[0])
fig, axes = plt.subplots(figsize=(20, 20), ncols = len(conditions))
for i in range(len(conditions)):
#     odomPercentile = vehicles[vehicles.odometer.between(quantiles[i], quantiles[i+1])]
    filteredData = vehicles[vehicles.condition.values == conditions[i]]
    prices = filteredData['price']
    axes[i].violinplot(prices.dropna(), showmeans = True)
    axes[i].set_title("condition: {}".format(conditions[i]))
```

```python
x = "odometer"
y = "price"

xQuantiles = [0]
data = {}
for i in range(10):
    xQuantiles.append(vehicles[x].quantile((i+1)/10))
for i in range(10):
    xQuantilesFrame = vehicles[vehicles.odometer.between(xQuantiles[i], xQuantiles[i+1])]
    yQuantiles = [0]
    yMeans = []
    for j in range(10):
        yQuantiles.append(xQuantilesFrame[y].quantile((j+1)/10))
    for j in range(10):
        yQuantilesFrame = xQuantilesFrame[xQuantilesFrame[y].between(yQuantiles[j], yQuantiles[j+1])]
        yMean = yQuantilesFrame[y].mean()
        yMeans.append(round(yMean))
    data["{}-{}".format(round(xQuantiles[i]), round(xQuantiles[i+1]))] = yMeans
percentileFrame = pd.DataFrame(data)
print(percentileFrame)
        
```
